```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def ploteo_complejo(mat,indicador,escala):
    if(indicador=="I"):
        mat=np.abs(mat)**2
    elif(indicador=="A"):
        mat=np.abs(mat)
    elif(indicador=="p"):
        mat=np.angle(mat)
    else:
        mat=mat
        
    if(escala==1):
        plt.figure()
        plt.imshow(mat)
        plt.show()
    else:
        plt.figure()
        plt.imshow(np.log(mat+0.00001))
        plt.show()
        
    return 0
```

```python
img=cv2.imread('mar.JPG',0)
fft_img=np.fft.fftn(img)
fft_img=np.fft.fftshift(fft_img)
```

```python
%matplotlib qt
```

```python
plt.figure()
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
mask =np.zeros(np.shape(img), dtype="uint8")
cv2.rectangle(mask,(50,150),(200,300),10,-1) # 10 valor de los pxeles en la mascara menos 1 significa que quiero que me rellene todo
plt.figure()
plt.imshow(mask)
plt.show()
```

```python
np.shape(img)
mask2=np.ones(np.shape(img),dtype="uint8")
coordx=int(np.shape(img)[1]/2)
coordy=int(np.shape(img)[0]/2) # pa que est en la mitad si la imagen es cuadrada
#print (coord)
cv2.circle(mask2,(coordx,coordy),40,0,-1)
plt.figure()
plt.imshow(mask2)
plt.colorbar()
plt.show ()
```

```python
new_img=np.fft.fftn(mask2*fft_img)
```

```python
plt.figure()
plt.imshow(np.abs(new_img)**2)
plt.show()
```

```python
ploteo_complejo(new_img,"I",1)
```
